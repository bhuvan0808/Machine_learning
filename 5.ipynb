```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("data.csv")
```

```python
X = df[["sMoothnessMean","periMeterMean"]]
kmeans = KMeans(n_clusters=10)
X["sMoothnessMean"] = kmeans.fit_predict(X)
X["sMoothnessMean"] = X["sMoothnessMean"].astype("category")
X.head()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
Z= linkage(X, 'ward')
```

```python
dendrogram (Z, leaf_rotation=90, leaf_font_size=5, labels=df.index)
```
