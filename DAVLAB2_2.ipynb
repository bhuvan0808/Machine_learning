17. Fill the null values of both categorical/text and numeric columns with the most frequent value(mode) by using sklearn library's imputer method on only specific rows and columns using iloc method

```python
import pandas as pd
import numpy as np







df3 = pd.read_csv('coin_gecko_2022-03-17.csv')
df3
```

```python
df3 = pd.read_csv('coin_gecko_2022-03-17.csv')

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
df3.iloc[:,:]=imputer.fit_transform(df3.iloc[:,:].values)
df3
```

```python
df3_selected = df3.iloc[:, [0, 1, 2]]

imputer = SimpleImputer(strategy='most_frequent')

imputer.fit(df3_selected)

df3_imputed = imputer.transform(df3_selected)

df3_selected = pd.DataFrame(df3_imputed, columns=df3_selected.columns)

# Print the result
print(df3_selected.head())
```
