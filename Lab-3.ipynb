# Q1: Write a python program to read from a file (sample_file.txt) and check if data is string or numbers, print and save only numbers into another file (numbers.txt)

```python
file1 = open('sample_file.txt', 'r')
file2 = open('numbers.txt', 'w')
x = file1.readlines()
for i in x:
    i = i.strip()
    if i.isdigit():
        file2.write(i + "\n")
file1.close()
```

# Q2: From Question 1, (numbers.txt) if number, check whether the number is an Armstrong numbers or not.If it's an armstrong number, save it in another file called 'armstrong_num.txt'

```python
def is_armstrong_number(num):
    num_str = str(num)
    num_digits = len(num_str)
    sum_of_powers = sum(int(digit)**num_digits for digit in num_str)
    return sum_of_powers == num
with open('numbers.txt', 'r') as infile, open('armstrong_num.txt', 'w') as outfile:
    for i in infile:
        i = i.strip()
        if i.isdigit():
            num = int(i)
            if is_armstrong_number(num):
                outfile.write(str(num) + '\n')
```
